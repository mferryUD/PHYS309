Plot Frank's measurement of the y-component of the proton velocity at the edge of the cone, as a function of spacecraft speed.</li>

```python
# Your code here:
c=constants.c
v=np.linspace(0.01,c,num=250,endpoint=False)
up=0.99*c
Beta=v/c
gma=1/((1-Beta**2)**0.5)
u=gma*((up+v)/(1+up*v/(c**2)))
plt.semilogy(v,u)
plt.xlabel('Speed of Spacecraft')
plt.ylabel('Speed of Proton rel to Space Station')
plt.show()

Ang=15
AngR=Ang*(np.pi/180)
uyp=np.sin(AngR)*up
uy=uyp/(gma*(1+up*v/(c**2)))
plt.plot(v,uy)
plt.xlabel('Speed of Spacecraft')
plt.ylabel('Speed of Proton y-component')
plt.show
```

Based only on your plot, would you say the flashes were simultaneous the point of view of the James Webb Space Telescope? What about from the point of view of the microprobe?

```python
# Your code here
t = np.linspace(0, 200, num=201, endpoint=True)
c=constants.c
fig, ax = plt.subplots(figsize=(7,7))

#Equation Set Up
#2A
xFB1pos=c*t
xFB1neg=-c*t

xFB2pos=c*t+3e8
xFB2neg=-c*t+3e8

#2B
xJW=1.5e8*np.ones(len(t))

#2C
xAMP=-0.1*c*t+5e8

#Plotting
#2A
ax.plot(xFB1pos,c*t,'g', label='Flashbulb @x=0')
ax.plot(xFB1neg,c*t,'g')
ax.plot(xFB2pos,c*t,'r', label='Flashbulb @x=3e8')
ax.plot(xFB2neg,c*t,'r')

#2B
ax.plot(xJW,c*t,'0.6', label='James Webb Space Telescope ')
ax.plot(1.5e8,0,'ro')

#2C
ax.plot(xAMP,c*t,'y', label='Alien Microprobe')
ax.plot(5e8,0,'ro')

#Setting up figure
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_aspect('equal')
ax.set_xlim([-1e8, 6e8])
ax.set_ylim([-3.7e8, 6e8])
plt.legend()
ax.text(6.1e8,0,'x')
ax.text(0,6.1e8,'ct')

#Question Answer
print('From the perspective of the James Webb Space Telescope, the two flashes happen simultaneously as seen by their world lines. From the perspective of the microprobe, the bulb at 3e8m appears to flash before the bulb at 0m.')
```
